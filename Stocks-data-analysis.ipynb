```python
import pandas as pd
dataframe = pd.read_csv('Stocks-data.csv')
dataframe.head()
```

```python
dataframe.set_index('證券名稱',inplace=True)
```

```python
dataframe.head()
```

```python
dataframe.dtypes
```

```python
def sum(row):
    sum = ""
    for item in row:
        sum += item
    return sum
```

```python
dataframe['成交股數'] = dataframe['成交股數'].map(lambda x:x.split(','))
dataframe['成交金額'] = dataframe['成交金額'].map(lambda x:x.split(','))
dataframe['成交筆數'] = dataframe['成交筆數'].map(lambda x:x.split(','))
dataframe['開盤價'] = dataframe['開盤價'].map(lambda x:x.split(','))
dataframe['最高價'] = dataframe['最高價'].map(lambda x:x.split(','))
dataframe['最低價'] = dataframe['最低價'].map(lambda x:x.split(','))
dataframe['收盤價'] = dataframe['收盤價'].map(lambda x:x.split(','))
dataframe['漲跌價差'] = dataframe['漲跌價差'].map(lambda x:x.split(','))
```

```python
dataframe['成交股數'] = dataframe['成交股數'].map(sum)
dataframe['成交金額'] = dataframe['成交金額'].map(sum)
dataframe['成交筆數'] = dataframe['成交筆數'].map(sum)
dataframe['開盤價'] = dataframe['開盤價'].map(sum)
dataframe['最高價'] = dataframe['最高價'].map(sum)
dataframe['最低價'] = dataframe['最低價'].map(sum)
dataframe['收盤價'] = dataframe['收盤價'].map(sum)
dataframe['漲跌價差'] = dataframe['漲跌價差'].map(sum)
```

```python
dataframe.head()
```

```python
dataframe['成交股數'] = dataframe['成交股數'].astype('int32')
dataframe['成交金額'] = dataframe['成交金額'].astype('int64')
dataframe['成交筆數'] = dataframe['成交筆數'].astype('int32')
dataframe['開盤價'] = dataframe['開盤價'].astype(float)
dataframe['最高價'] = dataframe['最高價'].astype(float)
dataframe['最低價'] = dataframe['最低價'].astype(float)
dataframe['收盤價'] = dataframe['收盤價'].astype(float)
```

```python
dataframe['平均'] = dataframe['成交金額']/dataframe['成交股數']
dataframe.head()
```

```python
dataframe2=dataframe.sort_values(by='成交股數',ascending=False)
dataframe2.head()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure(figsize=(20,10))
ax1 = figure.add_subplot(1,1,1)
ax1.plot(dataframe2['平均'],dataframe2['成交金額'])
plt.show()
```
